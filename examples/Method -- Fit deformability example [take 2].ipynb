```python
def reg_m(y, x):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    results = sm.OLS(y, X).fit()
    return results
```

```python
vmin = 0.93 # 0.934
vmax = 1.02 # 1.007

xlim_first = [2, 9.8]
ylim_first = [0.3, 1.2]

xlim = [2.3, 3.5]
ylim = [0.87, 1.12]

deviation=0.02
```

```python
def display_fit(width, eqPonA, factor, model, fig, ax, xlim=None, ylim=None):
#     y = np.arange(np.floor(np.min(width)) - 0.5, np.ceil(np.max(width)) + 0.5, 0.01)
#     x = np.arange(np.floor(np.min(eqPonA)*10)/10 - 0.2, np.ceil(np.max(eqPonA)*10)/10 + 0.2, 0.002)

    x = np.arange(xlim[0], xlim[1] + 0.05, 0.05)
    y = np.arange(ylim[0], ylim[1] + 0.01 , 0.01)

    xx, yy = np.meshgrid(x, y)

    zz = model(xx, yy)
    give = fit_give(
        xx, yy, width, eqPonA, factor, kx=2, ky=1, deviation=deviation)

    outOfTolerance = (give > 0.5)
    
#     valid_only = zz.copy()    
#     valid_only[outOfTolerance] = np.nan
    
#     print(np.nanmax(valid_only))
#     print(np.nanmin(valid_only))

#     c = ax.contourf(xx, yy, zz, 100, alpha=1, cmap=green_cm, vmin=vmin, vmax=vmax)
    ax.contourf(xx, yy, zz, 100, alpha=0.8, cmap=green_cm, vmin=vmin, vmax=vmax)
#     cb = plt.colorbar(
#         c, cax=cbar_ax, ticks=np.arange(vmin - 0.1, vmax + 0.1, 0.005),
#         orientation='horizontal')
    
    norm = Normalize(vmin=vmin, vmax=vmax)

    cb = ColorbarBase(
        cbar_ax, cmap=green_cm, alpha=0.8,
        norm=norm,
        orientation='horizontal')
    
    cb.ax.xaxis.set_ticks_position('top')
    cb.ax.xaxis.set_label_position('top')
    
    cb.set_clim([vmin, vmax])
    
#     ax.yaxis.set_ticks_position('right')
#     ax.yaxis.set_label_position('right')

    ax.contour(xx, yy, give, levels=np.linspace(0.498,0.502), colors='k')

    ax.scatter(width, eqPonA, s=50, alpha=0.9, 
                c='black', label=r'Measured data')

    ax.set_xlabel(r'Width (cm)')
    ax.set_ylabel(r'Perimeter / Area (cm$^{-1}$)')
```

```python
def display_fit_3D(width, eqPonA, factor, model, fig, ax, 
                   give_level=None, colour_wash=True,
                   top_cut=None, bot_cut=None):    
#     x = np.arange(xlim[0], xlim[1], 0.005)
#     y = np.arange(ylim[0], ylim[1], 0.001)


    x = np.linspace(xlim[0], xlim[1], 200)
    y = np.linspace(ylim[0], ylim[1], 200)

    xx, yy = np.meshgrid(x, y)

    zz = model(xx, yy)
    
#     if top_cut is not None:
# #         zz[zz>top_cut] = np.nan
#         zz[zz>top_cut] = top_cut
        
#     if bot_cut is not None:
# #         zz[zz<bot_cut] = np.nan
#         zz[zz<bot_cut] = bot_cut
    
    if give_level is not None:
        give = fit_give(
            xx, yy, width, eqPonA, factor, kx=2, ky=1, deviation=deviation)
        give = give + give_level - 0.5 
        ax.contour(xx, yy, give, levels=[give_level], colors='k')
    
    
#     small = (zz < 0)
#     zz[small] = 0
    
    if colour_wash:
        ax.plot_surface(xx, yy, zz, rstride=10, cstride=10, alpha=0.8,
                        cmap=green_cm, lw=0.2, vmin=vmin, vmax=vmax, zorder=-10)
    else:
        ax.plot_surface(xx, yy, zz, rstride=10, cstride=10, alpha=0.9,
                        lw=0.2, zorder=-10)
    
#     ax.scatter(width, eqPonA, factor, s=50, alpha=1, c='black')
    
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    ax.set_zlim([vmin, 0.96])

    ax.set_xlabel(r'Width (cm)')
    ax.set_ylabel(r'Perimeter / Area (cm$^{-1}$)')
    ax.zaxis.set_rotate_label(False)
    ax.set_zlabel(r'Insert factor', rotation=91.5)
```
