```python
from scipy.interpolate import UnivariateSpline
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,10)
y = np.sin(x)

plt.plot(x,y)
```

```python
cubic = UnivariateSpline(x, y, k=2, s=0.4)
x_test = np.linspace(0, 10)
cubic_y = cubic(x_test)
```

```python
plt.plot(x_test, cubic_y)
plt.scatter(x,y)
```
