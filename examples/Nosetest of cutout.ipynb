```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import shapely.affinity as aff

import sys
sys.path.append("../")
from electronfactors.ellipse.equivalent import equivalent_ellipse
from electronfactors.visuals.shape_display import display_equivalent_ellipse, display_shapely

from electronfactors.ellipse.utilities import shapely_cutout, shapely_ellipse
```

```python
XCoords = np.array([-1, -0.2, 0, 0.7, 1, 0])*4
YCoords = np.array([0, -1, -.8, 0, .6, 1])*4

result = equivalent_ellipse(
    XCoords=XCoords, YCoords=YCoords)
```

```python
cutout = shapely_cutout(XCoords, YCoords)
```

```python
poi = result['poi']
width = result['width']
length = result['length']

ellipse = shapely_ellipse([poi[0], poi[1], width, length, 0])
ellipse = aff.rotate(ellipse, -42)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

display_shapely(ellipse, ax=ax)
display_shapely(cutout, ax=ax)
```
